### convert msisensor scan list to MANTIS input bed

```python
def convert2bed(msisensor_scan_result, mantis_input_bed):
    with open(msisensor_scan_result) as f, open(mantis_input_bed, 'w') as fw:
        for line in f:
            if line.startswith('chromosome'):
                continue
            its = line.strip().split('\t')
            chrom = its[0]
            start = int(its[1])
            unit = int(its[2])
            times = int(its[4])
            repeat_unit_bases, left_flank_bases, right_flank_bases = its[-3:]
            end = start + unit * times
            feature = "({}){}".format(repeat_unit_bases, times)
            unused1 = '.'
            unused2 = '{}-{}-{}'.format(left_flank_bases, feature, right_flank_bases)
            new_line = '\t'.join(str(x) for x in [chrom, start, end, feature, unused1, unused2])
            fw.write(new_line + '\n')
            print(new_line)
```

### JS散度+信息熵 优化msi

#### 1.信息熵


```python
def get_entropy(p, q):
    p = [i * 1.0 / sum(p) for i in p]
    q = [i * 1.0 / sum(q) for i in q]
    #if len(p) != 1:
        #p, q = zip(*filter(lambda x: x[0]!=0 or x[1]!=0, zip(p, q)))
    try: #remove both 0 col
        p, q = zip(*filter(lambda x: x[0]!=0 or x[1]!=0, zip(p, q)))
    except ValueError:
        print(p, q)
    p = p + np.spacing(1)
    q = q + np.spacing(1)
    e_p = sum([- i * log(i, 2) for i in p])
    e_q = sum([- j * log(j, 2) for j in q])
    return (e_p, e_q)
```

#### 2.JS散度

```python
#读取mantis结果
from collections import OrderedDict
def parse_count(infile):
    d = OrderedDict()
    with open(infile) as f:
        for line in f:
            if line.startswith('Locus'):
                continue
            locus, repeat, normal, tumor = line.strip().split('\t')
            #print(locus, repeat, normal, tumor)
            #break
            repeat, normal, tumor = [int(x) for x in [repeat, normal, tumor]]
            if locus not in d:
                d[locus] = {'normal': [normal], 'tumor': [tumor], 'repeat': [repeat]}
            else:
                d[locus]['normal'].append(normal)
                d[locus]['tumor'].append(tumor)
                d[locus]['repeat'].append(repeat)
    return d
```

##### KL散度和JS散度函数

```python
import numpy as np
KLD=lambda p, q: sum([_p * log(_p, 2)- _p * log(_q, 2) for (_p, _q) in zip(p, q)])
def JSD(p, q):
    bug = False
    p = [i * 1.0 / sum(p) for i in p]
    q = [i * 1.0 / sum(q) for i in q]
    #if len(p) != 1:
        #p, q = zip(*filter(lambda x: x[0]!=0 or x[1]!=0, zip(p, q)))
    try:
        p, q = zip(*filter(lambda x: x[0]!=0 or x[1]!=0, zip(p, q)))
    except ValueError:
        bug = True
        #print(p, q)
    m = [0.5 * (_p + _q) for _p, _q in zip(p, q)]
    p = p + np.spacing(1)
    q = q + np.spacing(1)
    m = m + np.spacing(1)
    return (0.5 * KLD(p, m) + 0.5 * KLD(q, m), bug)
```

```python
def cal_mean_jsd(infile):
    d_count = parse_count(infile)
    sum_jsd = 0
    cnt_filter_by_entropy = 0
    for locus in d_count:
        normals = d_count[locus]['normal']
        tumors = d_count[locus]['tumor']
        jsd, bug = JSD(normals, tumors)
        jsd = round(jsd, 4)
        if bug:
            print('[ERROR] {}'.format(locus))
            print(normals, tumors)
        sqrt_jsd = sqrt(jsd)
        #if sqrt_jsd > 0.3:
            #print(locus, sqrt_jsd)
        e_normal, e_tumor = get_entropy(normals, tumors)
        #print(e_normal, e_tumor)
        if e_normal > e_tumor:
            cnt_filter_by_entropy += 1
            continue
        sum_jsd += sqrt_jsd
    return (round(sum_jsd / len(d_count), 4), cnt_filter_by_entropy, len(d_count))
```
